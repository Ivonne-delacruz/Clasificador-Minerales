```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
classes = dict()

with open("list_image_urls.csv") as file:
    lines = file.readlines()
    for line in lines:
        label = line.split(",")[1].strip()
        if label in classes:
            classes[label] += 1
        else:
            classes[label] = 1
```

```python
class_list = list(classes.items())
class_list = sorted(class_list, key=lambda tup: tup[1], reverse=False)
```

```python
index = 0
max_points = class_list[-1][1]
classes_with_at_least_i_points = []
for i in range(1, max_points + 1):
    for k in range(index, len(classes)):
        if class_list[k][1] < i:
            index += 1
        else:
            classes_with_at_least_i_points.append(index)
            break
classes_with_at_least_i_points = [len(class_list) - c for c in classes_with_at_least_i_points]
```

```python
required = 500
tolerance = 2000
```

```python
plt.scatter(range(required, max_points + 1 - (max_points - tolerance)), classes_with_at_least_i_points[required-1:-(max_points - tolerance)], s=5)
plt.show()
```
