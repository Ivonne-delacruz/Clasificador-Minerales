### ***Número de Imágenes***


```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Se enlistan todas las categorías disponibles y el número de imágenes que tienen
classes = dict()

with open("url_list.csv") as file:
    lines = file.readlines()
    for line in lines:
        label = line.split(",")[1].strip()
        if label in classes:
            classes[label] += 1
        else:
            classes[label] = 1
```

```python
# Se ordenan las categorías de la más pequeña a la más grande
class_list = list(classes.items())
class_list = sorted(class_list, key=lambda tup: tup[1], reverse=False)
```

```python
# Análisis del número de categorías respecto al número de ejemplos
index = 0
max_points = class_list[-1][1]
classes_with_at_least_i_points = []
for i in range(1, max_points + 1):
    for k in range(index, len(classes)):
        if class_list[k][1] < i:
            index += 1
        else:
            classes_with_at_least_i_points.append(index)
            break
classes_with_at_least_i_points = [len(class_list) - c for c in classes_with_at_least_i_points]
```

```python
# Ventana de oportunidad
required = 500
filter = 5000
```

```python
# Número de categorías con al menos 8000 ejemplos
print(f"Hay {classes_with_at_least_i_points[5000]} clases con al menos {filter} ejemplos")
```

```python
# Listado de las categorías más grandes
for tup in class_list[-14:]:
    print(f"{tup[0].strip("\'")}: {tup[1]} imágenes")
```

```python
# Visualización
plt.figure()
plt.title("Número de categorías con al menos x ejemplos", fontsize=12)
plt.xlabel("Ejemplos", fontsize=10)
plt.ylabel("Categorías", fontsize=10)

plt.scatter(range(required, max_points + 1 - (max_points - filter)), classes_with_at_least_i_points[required-1:-(max_points - filter)], s=1, c="r")
plt.show()
```

### ***Tamaño de imágenes***

```python
import requests
from PIL import Image
from io import BytesIO
```

```python
# Engañando al servidor
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) '
                  'Chrome/91.0.4472.124 Safari/537.36 '
}
```

```python
# Obtiene el tamaño de una imagen
def get_image_size(image_url):
    try:
        response = requests.get(image_url, headers=headers, stream=True)
        response.raise_for_status()
        img = Image.open(BytesIO(response.content))
        return img.size
    except Exception as e:
        pass
        return None, None
```

```python
# Tamaño de las imágenes disponibles

widths = []
heights = []

# Selección de muestra
counter_from = 1
counter_to = 1000

# Lectura de la lista de urls
with open("filtered_url_list.csv") as file:
    lines = file.readlines()
    line_count = 0
    for line in lines:
        line_count += 1
        if line_count < counter_from:
            pass
        else:
            if line_count > counter_to:
                break
            url, label = line.strip().split(",")
            width, height = get_image_size(url)
            if width is not None and height is not None:
                widths.append(width)
                heights.append(height)
```

```python
# Visualización
plt.figure()
plt.title("Tamaños de imagen en el conjunto de datos", fontsize=12)

plt.xlabel("Ancho", fontsize=10)
plt.ylabel("Alto", fontsize=10)

plt.scatter(widths, heights, s=4, c="r")
plt.show()
```

### ***Imágenes de muestra***

```python
import os
import re
```

```python
# Descarga una imagen con la etiqueta como nombre
def download_image(image_url, image_label, download_folder, labels_dict):
    try:
        # Solicitud
        response = requests.get(image_url, headers=headers, stream=True)
        response.raise_for_status()
        img = Image.open(BytesIO(response.content))

        # Crea el nombre del archivo según su etiqueta
        img_extension = img.format.lower()
        file_name = f"{image_label}_{labels_dict[image_label]}.{img_extension}"
        file_path = os.path.join(download_folder, file_name)

        # Guarda la imagen en el directorio especificado
        img.save(file_path)
        # Pasa a la siguiente imagen
        labels_dict[image_label] += 1
    except Exception as e:
        print(f"Error al descargar la imagen desde {image_url}: {e}")
```

```python
# Lectura de la lista de clases seleccionadas
labels = dict()
with open("list_selected_classes (full_names).txt") as file:
    lines = file.readlines()
    for line in lines:
        label = line.strip()
        match = re.search(r"\(Var: ([^\)]+)\)", label)
        if match:
            label = match.group(1)
        labels[label] = 0
```

```python
# Directorio de descarga
download_dir = os.path.join(os.path.dirname(os.getcwd()), "Clasificador-Minerales", "images_samples")

if not os.path.exists(download_dir):
    os.makedirs(download_dir)
```

```python
# Guarda las imágenes de muestra

# Selección de muestra
counter_from = 11
counter_to = 19

# Descarga
with open("filtered_url_list.csv") as file:
    lines = file.readlines()
    line_count = 0
    for line in lines:
        line_count += 1
        if line_count < counter_from:
            pass
        else:
            if line_count > counter_to:
                break
            url, label = line.strip().split(",")
            download_image(url, label, download_dir, labels)
```

```python
import cv2
import glob
import numpy as np
```

```python
# Lista de rutas de las imágenes de muestra
paths = []
image_extensions = ["*.jpeg", "*.jpg", "*.png", "*.bmp", "*.mpo"]
for extension in image_extensions:
    paths.extend(glob.glob(os.path.join(download_dir, extension)))
```

```python
# Visualización
image_samples = [cv2.cvtColor(cv2.imread(path), cv2.COLOR_BGR2RGB) for path in paths]

fig, axs = plt.subplots(3, 3, sharex=True, sharey=True)
fig.suptitle("Imágenes de muestra", fontsize=12)

i=0
for ax in axs.flat:
    image_label = paths[i].split("\\")[-1].split("_")[0]
    ax.imshow(image_samples[i], aspect="equal")
    ax.set_title(image_label, fontsize=10, x=0.25, y=-0.25)
    ax.axis("off")
    i += 1

plt.show()
```

### ***Procesamiento de Imágenes***

```python
# Muestra una imagen
def show_image(image, title):
    plt.figure()
    plt.suptitle(title)
    plt.imshow(image)
    plt.axis("off")
    plt.show()   
```

```python
# Imagen original
sample_raw = image_samples[8]
show_image(sample_raw, "Imagen original")
```

```python
# Redimensiona una imagen
def resize(image, size):
    # Tamaño original de la imagen
    height, width, channels = image.shape

    # Orientación de la imagen
    if height < width:
        crop_width = (width - height) // 2
        image = image[:, crop_width:crop_width + height, :]
    else:
        crop_height = (height - width) // 2
        image = image[crop_height:crop_height + width, :, :]
    resized_image = cv2.resize(image, size, interpolation=cv2.INTER_LANCZOS4)
    return resized_image
```

```python
# Redimensionamiento
desired_size = (256, 256)
sample_resized = resize(sample_raw, desired_size)

show_image(sample_resized, "Imagen redimensionada")
```

```python
# Encuentra los bordes del mineral
def canny_filter(image, threshold1, threshold2):
    image = cv2.cvtColor(np.float32(image), cv2.COLOR_BGR2RGB)
    return cv2.Canny(image=cv2.convertScaleAbs(image), threshold1=threshold1, threshold2=threshold2)
```

```python
# Detección de bordes
canny_threshold_1 = 100
canny_threshold_2 = 75

sample_edges = canny_filter(sample_resized, canny_threshold_1, canny_threshold_2)

show_image(sample_edges, "Bordes detectados")
```

```python
# Desenfoque gaussiano aplicado a los bordes
gaussian_blur_kernel_size = (5, 5)
sample_blurred_edges = cv2.GaussianBlur(np.float32(sample_edges), gaussian_blur_kernel_size, 0)

show_image(sample_blurred_edges, "Bordes difuminados")
```

```python
# Rellena el contorno encontrado en la imagen
def fill_contours(image):
    contours, _ = cv2.findContours(image.astype(np.uint8), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    filled_image = np.zeros_like(image).astype(np.uint8)
    cv2.drawContours(filled_image, contours, -1, 1, thickness=cv2.FILLED)
    return filled_image
```

```python
# Máscara de relleno
sample_interior = fill_contours(sample_blurred_edges)

show_image(sample_interior, "Interior del mineral")
```

```python
# Elimina pequeñas islas del fondo, rescatando trozos del mineral
def fill_holes(image, foreground, connectivity, area_threshold, variance_threshold):
    # Máscara
    foreground_mask = foreground == 1
    # Inversión de valores
    background_mask = ~foreground_mask
    # Etiqueta las regiones conectadas
    num_labels, labels, stats, _ = cv2.connectedComponentsWithStats(background_mask.astype(np.uint8),
                                                                    connectivity=connectivity)
    # Valores promedios encontrados
    foreground_mean = np.mean(image[foreground_mask], axis=0)
    background_mean = np.mean(image[background_mask], axis=0)
    # Recorre cada componente
    for i in range(1, num_labels):
        area = stats[i, cv2.CC_STAT_AREA]
        region_mask = labels == i
        region_mean = np.mean(image[region_mask], axis=0)
        # Se rescata la isla si tiene un color más parecido al del mineral
        norms = [np.linalg.norm(background_mean-region_mean),
                 np.linalg.norm([255, 255, 255]-region_mean),
                 np.linalg.norm([0, 0, 0]-region_mean)]
        var = np.var(region_mean)
        if var > variance_threshold and area < area_threshold \
                and np.linalg.norm(foreground_mean - region_mean) < min(norms):
            foreground[region_mask] = 1

    return foreground
```

```python
# Criterio de conexidad
connectivity_diagonal = 8
# Criterio de eliminación de agujeros
area_threshold_sup = (desired_size[0] * desired_size[1]) // 3
# Varianza
variance_threshold = 0.01

sample_filled = fill_holes(sample_resized, sample_interior.copy(), connectivity_diagonal, area_threshold_sup, variance_threshold)

show_image(sample_filled, "Fondo rellenado")
```

```python
# Elimina pequeñas islas del mineral
def remove_components(image, connectivity, threshold):
    # Etiqueta las regiones conectadas
    num_labels, labels, stats, _ = cv2.connectedComponentsWithStats(image, connectivity=connectivity)
    # Recorre cada componente
    for i in range(1, num_labels):
        area = stats[i, cv2.CC_STAT_AREA]
        # Se elimina la isla si su área es menor que el umbral especificado
        if area < threshold:
            image[labels == i] = 0
    return image
```

```python
# Criterio de conexidad
connectivity_simple = 4
# Criterio de eliminación de islas
area_threshold_inf = (desired_size[0] * desired_size[1]) // 32

sample_cleaned = remove_components(sample_filled.copy(), connectivity_simple, area_threshold_inf)

show_image(sample_cleaned, "Fondo limpio")
```

```python
# Imagen procesada
sample_processed = np.round(sample_resized * sample_cleaned[:, :, np.newaxis]).astype(np.uint8)

show_image(sample_processed, "Imagen procesada")
```

```python
# Redimensionado para la red neuronal
input_size = (96, 96)

sample_input=cv2.resize(sample_processed, input_size, interpolation=cv2.INTER_LANCZOS4)

show_image(sample_input, "Imagen usada en el modelo")
```

```python
# Visualización
sample_history = [sample_raw, sample_resized, sample_edges, 
                  sample_blurred_edges, sample_interior, sample_filled, 
                  sample_cleaned, sample_processed, sample_input]

fig, axs = plt.subplots(1, 9, figsize=(14, 2), gridspec_kw={'width_ratios': [3, 1, 1, 1, 1, 1, 1, 1, 1]})
fig.suptitle("Flujo de procesamiento", fontsize=12)

i=0
for ax in axs.flat:
    ax.imshow(sample_history[i])
    ax.axis("off")
    i += 1

plt.show()
```

```python
# Flujo de Procesamiento
def process_image(image):
    resized_image = resize(image, desired_size)
    return cv2.resize(np.round(
                        resized_image * remove_components(fill_holes(
                                                                     resized_image, fill_contours(cv2.GaussianBlur(np.float32(canny_filter(
                                                                                                                                           resized_image, canny_threshold_1, canny_threshold_2
                                                                                                                                          )
                                                                                                                             ), 
                                                                                                                   gaussian_blur_kernel_size, 0
                                                                                                                  )
                                                                                                 ), 
                                                                     connectivity_diagonal, area_threshold_sup, variance_threshold
                                                                    ), 
                                                          connectivity_simple, area_threshold_inf
                                                         )[:, :, np.newaxis]
                       ).astype(np.uint8), input_size, interpolation=cv2.INTER_LANCZOS4
              )
```

```python
# Visualización
processed_samples = [process_image(sample) for sample in image_samples]

fig, axs = plt.subplots(3, 3, sharex=True, sharey=True)
fig.suptitle("Imágenes procesadas", fontsize=12)

i=0
for ax in axs.flat:
    image_label = paths[i].split("\\")[-1].split("_")[0]
    ax.imshow(processed_samples[i])
    ax.set_title(image_label, fontsize=10, x=0.5, y=-0.2)
    ax.axis("off")
    i += 1

plt.show()
```
